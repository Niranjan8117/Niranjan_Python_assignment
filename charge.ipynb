```python
import matplotlib.pyplot as plt
import numpy as np

voltage = np.linspace(3.2, 4.2, 100)
cycle_counts = [1, 50, 100, 150]

ica_data = {
    1: np.random.normal(0, 5, 100) + 20 * np.exp(-(voltage - 3.7)**2 / 0.01),
    50: np.random.normal(0, 5, 100) + 18 * np.exp(-(voltage - 3.75)**2 / 0.012),
    100: np.random.normal(0, 5, 100) + 15 * np.exp(-(voltage - 3.8)**2 / 0.015),
    150: np.random.normal(0, 5, 100) + 12 * np.exp(-(voltage - 3.85)**2 / 0.018),
}


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for cycle_count in cycle_counts:
    ax.plot(voltage, np.full(len(voltage), cycle_count), ica_data[cycle_count])

ax.set_xlabel('Voltage (V)')
ax.set_ylabel('Cycle Count (Aging)')
ax.set_zlabel('Incremental Capacity (dQ/dV)')
ax.set_title('Incremental Capacity vs. Voltage and Aging')

plt.show()
```
